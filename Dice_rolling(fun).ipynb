```python
import random

# Dice faces using ASCII art
dice_faces = {
    1: (
        "┌─────┐\n"
        "│     │\n"
        "│  ●  │\n"
        "│     │\n"
        "└─────┘"
    ),
    2: (
        "┌─────┐\n"
        "│ ●   │\n"
        "│     │\n"
        "│   ● │\n"
        "└─────┘"
    ),
    3: (
        "┌─────┐\n"
        "│ ●   │\n"
        "│  ●  │\n"
        "│   ● │\n"
        "└─────┘"
    ),
    4: (
        "┌─────┐\n"
        "│ ● ● │\n"
        "│     │\n"
        "│ ● ● │\n"
        "└─────┘"
    ),
    5: (
        "┌─────┐\n"
        "│ ● ● │\n"
        "│  ●  │\n"
        "│ ● ● │\n"
        "└─────┘"
    ),
    6: (
        "┌─────┐\n"
        "│ ● ● │\n"
        "│ ● ● │\n"
        "│ ● ● │\n"
        "└─────┘"
    )
}

def roll_dice(num_dice):
    return [random.randint(1, 6) for _ in range(num_dice)]

def show_dice(results):
    for value in results:
        print(dice_faces[value])
        print()  # space between dice

def main():
    history = []

    print("🎲 Dice Rolling Simulator 🎲")

    while True:
        try:
            num = int(input("\nHow many dice? "))
            if num <= 0:
                print("Enter a positive number.")
                continue
        except ValueError:
            print("Enter a valid number.")
            continue

        result = roll_dice(num)
        total = sum(result)

        print("\n🎲 Your Dice:")
        show_dice(result)

        print(f"Total: {total}")

        history.append((result, total))

        choice = input("\nRoll again? (y/n): ").lower()
        if choice != 'y':
            break

    print("\n📜 History:")
    for i, (res, tot) in enumerate(history, 1):
        print(f"{i}. {res} → {tot}")

main()
```
