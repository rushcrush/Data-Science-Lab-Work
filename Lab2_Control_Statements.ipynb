```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Data Access
The dataset was released in three file formats: CSV, TXT, and Excel. To check if different file formats affect the data understanding, each file was brought into a pandas dataframe.

Though all the files had an equal number of entries, Python treated them differently. The CSV file triggered syntax errors because of inconsistent use of delimiters. The TXT file, on the other hand, gained a few more unnamed columns due to the spacing issue. The Excel file nicely maintained both the dataset structure and the data types.

Hence, even though the data were the same, the file format had an impact on how correct the dataset was imported. Among them, the Excel file turned to be the most trustworthy data source.

```python
df_csv = pd.read_csv("Facebook Friends.csv")
df_csv.head()
```

```python
df_txt = pd.read_csv("Facebook Friends.txt", sep="\t", encoding="utf-16")
df_txt.head()
```

```python
df_xlsx = pd.read_excel("Facebook Friends.xlsx")
df_xlsx.head()
```

## 2. Data Examination

To understand the dataset's structure, pandas functions were used to check the Excel file.

Firstly, the head() and tail() functions were employed to print the first and last records, respectively, thus verifying that the dataset had been loaded correctly. By using the columns function, 18 variables were identified that outline the user demographics and Facebook activity, e.g., age, posts, replies, likes, and number of friends.

The info() function revealed that there are 715 observations in total, and all variables are stored as integer data types. With the help of the isnull().sum() function, it was found that there are no missing values in any of the columns. This means the dataset is complete and can be used for analysis without any cleaning.

# 3. Descriptive Statistics
Descriptive statistics were obtained by using the describe() function which gave us several statistics such as mean, standard deviation, min, max, and quartiles.

The comparison of the mean and the median values reveals the existence of outliers in the variables posts, replies, and friends. In fact, while the majority of users exhibit moderate levels of activity, a few users demonstrate extremely high levels of engagement.

The standard deviation figures reveal that there is a wide range of user behavior. On one hand, there are users who hardly ever interact and, on the other hand, there are users who post and reply very frequently. In general, Facebook activity is very unbalanced with respect to different users.

# 4. Data Visualization
 Histograms and other frequency charts have been used to depict the behavior of the users visually.

The age histogram displays that the majority of users are from the younger age bracket. The friends distribution being right, skewed indicates that the main part of the users has a moderate number of friends and only a few users have very large networks.

The posts histogram reveals that the level of activity is uneven, as the small group of users contributes the bulk of the posts. The gender frequency plot presents the distribution of users among different categories.

Visualization helps to uncover the main patterns and outliers and thus it is a good supplement to descriptive statistics.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=15)
plt.title("Age Distribution of Facebook Users")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Friends"], bins=20)
plt.title("Distribution of Facebook Friends")
plt.xlabel("Number of Friends")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Posts"], bins=20)
plt.title("Distribution of User Posts")
plt.xlabel("Number of Posts")
plt.ylabel("Frequency")
plt.show()
```

```python
df.columns = df.columns.str.strip()
```

```python
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.xlabel("Gender (0 = Male, 1 = Female)")
plt.ylabel("Count")
plt.show()
```
